```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
Sales_data = pd.read_csv('https://raw.githubusercontent.com/Tiamiyu1/Python-for-Data-Science-and-Analytics/main/Data/store_data.csv')
Sales_data
```

Observation: There are 200 rows and 6 columns in the sales dataset.

```python
Sales_data['week'] = pd.to_datetime(Sales_data['week'])
Sales_data['year'] = Sales_data['week'].dt.year
Sales_data['month'] = Sales_data['week'].dt.month
Sales_data['day'] = Sales_data['week'].dt.day
Sales_data
```
